```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter
from langchain_google_vertexai import VertexAIEmbeddings
from langchain.embeddings import HuggingFaceEmbeddings
from langchain_chroma import Chroma
```

```python
from dotenv import load_dotenv

load_dotenv()
```

```python
import pandas as pd

books = pd.read_csv('books_cleaned.csv')
books
```

```python
books["tagged_description"]
```

```python
books['tagged_description'].to_csv("tagged_descriptions.txt", sep="\n", index=False, header=False)
```

```python
raw_documents = TextLoader("tagged_descriptions.txt").load()
text_splitter = CharacterTextSplitter(chunk_size=0, chunk_overlap=0, separator="\n")
documents = text_splitter.split_documents(raw_documents)
```

```python
documents[0]
```

```python
embedding = HuggingFaceEmbeddings(model_name='paraphrase-MiniLM-L6-v2')

db_books = Chroma.from_documents(
    documents,
    embedding=embedding
)
```

```python
query = "A book to teach children about nature"
docs = db_books.similarity_search(query, k=10)
docs
```

```python
books[books["isbn13"] == int(docs[0].page_content.split()[0].strip())]
```

```python
def retrieve_semantic_recommendation(
        query: str,
        top_k: int = 10,
) -> pd.DataFrame:
    recs = db_books.similarity_search(query, k=top_k)

    books_list = []

    for i in range(0, len(recs)):
        books_list += [int(recs[i].page_content.strip('"').split()[0])]

    return books[books["isbn13"].isin(books_list)].head(top_k)
```

```python
retrieve_semantic_recommendation("A heartwarming juerny of love and firndship")
```
